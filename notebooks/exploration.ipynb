# PHENOCAST Exploratory Analysis

Quick exploration of the StudentLife feature distributions and label balance.
Run `src/feature_engineering.py` before this notebook.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

X = pd.read_csv('../data/processed/features.csv', index_col='uid')
y = pd.read_csv('../data/processed/labels.csv', index_col='uid')['elevated']
print(f'n={len(X)}, positives={y.sum()}, prevalence={y.mean():.1%}')
```

```python
# Feature distributions split by label
X_pos = X[y == 1]
X_neg = X[y == 0]
print('Elevated PHQ-9 group (n=%d):' % len(X_pos))
print(X_pos.describe().T[['mean','std']].head(10))
print()
print('Non-elevated group (n=%d):' % len(X_neg))
print(X_neg.describe().T[['mean','std']].head(10))
```

```python
# Conversation feature: largest gap between groups
conv_cols = [c for c in X.columns if 'conversation' in c]
if conv_cols:
    fig, axes = plt.subplots(1, len(conv_cols), figsize=(4*len(conv_cols), 3))
    if len(conv_cols) == 1: axes = [axes]
    for ax, col in zip(axes, conv_cols):
        ax.hist(X_neg[col].dropna(), alpha=0.6, label='Non-elevated', bins=15)
        ax.hist(X_pos[col].dropna(), alpha=0.6, label='Elevated', bins=15)
        ax.set_title(col)
        ax.legend()
    plt.tight_layout()
    plt.show()
```
